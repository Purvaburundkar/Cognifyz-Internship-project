**Task 1 - Table Booking and Online Delivery**

```python
data.head()
```

```python
#Converting Has Table booking column tu numeric

from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

data['Has Table booking']= label_encoder.fit_transform(data['Has Table booking'])

data['Has Table booking'].unique()
```

```python
#Converting Has Online delivery column tu numeric

label_encoder = preprocessing.LabelEncoder()

data['Has Online delivery']= label_encoder.fit_transform(data['Has Online delivery'])

data['Has Online delivery'].unique()
```

```python
data.head()
```

```python
table_booking_percentage = data['Has Table booking'].mean() * 100
online_delivery_percentage = data['Has Online delivery'].mean() * 100

print(f"Percentage of restaurants with table booking: {table_booking_percentage:.2f}%")
print(f"Percentage of restaurants with online delivery: {online_delivery_percentage:.2f}%")
```

```python
plt.figure(figsize=(13, 4))
plt.subplot(1, 2, 1)
sns.countplot(x='Price range', hue='Has Table booking', data=data)
plt.title(' Price Range Vs Online Delivery')


plt.subplot(1, 2, 2)
sns.countplot(x='Price range', hue='Has Online delivery', data=data)
plt.title(' Price Range Vs Table Booking')

plt.show()
```

```python
#Compareing the average ratings of restaurants with table booking and those without

table_booking_ratings = data[data['Has Table booking'] == 1]['Aggregate rating']
no_table_booking_ratings = data[data['Has Table booking'] == 0]['Aggregate rating']

print(f"Average rating of restaurants with table booking: {table_booking_ratings.mean():.2f}")
print(f"Average rating of restaurants without table booking: {no_table_booking_ratings.mean():.2f}")
```

```python
#Analyzing  the availability of online delivery among restaurants with different price ranges
price_ranges = data['Price range'].unique()

plt.figure(figsize=(12, 6))
for price_range in price_ranges:
  restaurants_with_online_delivery = data[data['Price range'] == price_range]['Has Online delivery']
  online_delivery_percentage = restaurants_with_online_delivery.mean() * 100
  plt.bar(price_range, online_delivery_percentage, label=f"{price_range} ({online_delivery_percentage:.2f}%)")

plt.xlabel("Price Range")
plt.ylabel("Percentage of Restaurants with Online Delivery")
plt.title("Availability of Online Delivery Among Restaurants with Different Price Ranges")
plt.legend()
plt.show()
```

**Task 2 - Price Range Analysis**

```python
#Determineing the most common price range among all the restaurants

most_common_price_range = data['Price range'].value_counts().idxmax()
print(f"Most common price range: {most_common_price_range}")
```

```python
#Calculating the average rating for each price range.

price_range_ratings = {}
for price_range in price_ranges:
  restaurants_in_price_range = data[data['Price range'] == price_range]['Aggregate rating']
  average_rating = restaurants_in_price_range.mean()
  price_range_ratings[price_range] = average_rating

print(f"Average rating for each price range:")
print(price_range_ratings)
```

```python
#Identify the color that represents the highest average rating among different price ranges using rating colour column

price_range_ratings = {}
for price_range in price_ranges:
  restaurants_in_price_range = data[data['Price range'] == price_range]['Aggregate rating']
  average_rating = restaurants_in_price_range.mean()
  price_range_ratings[price_range] = average_rating

highest_rating = max(price_range_ratings.values())
highest_rating_price_range = [k for k, v in price_range_ratings.items() if v == highest_rating]

print(f"The highest average rating is {highest_rating} and it is associated with the following price ranges:")
print(highest_rating_price_range)

highest_rating_color = data[data['Price range'].isin(highest_rating_price_range)]['Rating color'].unique()[0]

print(f"The color that represents the highest average rating is {highest_rating_color}")
```
